```python
action = REC.action_menu.clicked
func = REC.action_dict[action]["func"]
cells = REC.grid_row.objects[0][1].selection
REC.recordings_object.dataframes[
    "cell_selection"
] = REC.recordings_object.dataframes[REC.dataframe_menu.clicked].iloc[cells]
stimulus_indices = (
    REC.recordings_object.dataframes["cell_selection"]
    .stimulus_index.unique()
    .tolist()
)
spikes = REC.recordings_object.get_spikes_df("cell_selection")
fig = func(df=spikes)

flash_duration = stimulus_spikes.mean_trigger_times(
    REC.recordings_object.dataframes["stimulus_df"], stimulus_indices
)
fig = REC.ct.add_stimulus_to_plot(fig, flash_duration)
```
